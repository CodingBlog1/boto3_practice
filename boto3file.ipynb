```python

import boto3
import pandas as pd
from io import BytesIO
```

```python
file_path = 'Stapled_Booklets1.csv'
session = boto3.Session(
    aws_access_key_id='',
    aws_secret_access_key='',
)
s3 = session.resource('s3')

s3.meta.client.upload_file(Filename=file_path, Bucket='testbucketsoyab', Key='Booklets1.csv')
```

```python
#read csv file from s3
obj = s3.Bucket('testbucketsoyab').Object('Booklets1.csv').get()
csv_bytes = obj['Body'].read()
```

```python
#download file from s3
s3.Bucket('testbucketsoyab').download_file('Booklets1.csv','test.csv')
```

```python
try:
    df = pd.read_csv(BytesIO(csv_bytes), encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv(BytesIO(csv_bytes), encoding='iso-8859-1')

df
```
